```python
import numpy as np
import pandas as pd
```

```python
# Create a dataframe of ten rows, four columns with random values. Write a Pandas program to highlight the negative numbers red and positive numbers black.

np.random.seed(20)
data = pd.DataFrame({'A' : np.linspace(1,10,10)})
data = pd.concat([data, pd.DataFrame(np.random.randn(10,4), columns = list ('BCDE'))], axis = 1)
print(data)

def color_negative_red(num):
  if num < 0:
    color = 'red',
  else:
    color = 'black'
  return 'color: %s' % color

print("Negative Numbers Red and Positive Numbers Black")
data.style.applymap(color_negative_red)
```

```python
#Create a dataframe of ten rows, four columns with random values. Convert some values to nan values. Write a Pandas program which will highlight the nan values.
np.random.seed(20)
data = pd.DataFrame({'A' : np.linspace(1,10,10)})
data = pd.concat([data, pd.DataFrame(np.random.randn(10,4), columns = list('BCDE'))], axis = 1)
data.iloc[0,2] = np.nan
data.iloc[1,3] = np.nan
data.iloc[4,4] = np.nan
data.iloc[8,3] = np.nan
print(data)

print("Null value with Red color")
data.style.highlight_null(null_color='red')
```

```python
# Create a dataframe of ten rows, four columns with random values. Write a Pandas program to highlight the maximum value in each column.
# Create a dataframe of ten rows, four columns with random values. Write a Pandas program to highlight the minimum value in each column.

np.random.seed(20)
data = pd.DataFrame({'A' : np.linspace(1,10,10)})
data = pd.concat([data, pd.DataFrame(np.random.randn(10,4), columns = list('BCDE'))], axis = 1)
data.iloc[0,2] = np.nan
data.iloc[1,3] = np.nan
data.iloc[4,4] = np.nan
data.iloc[8,3] = np.nan


def highlight_max(num):
  is_max = num == num.max()
  return ['background-color: green' if i else " " for i in is_max]


def highlight_min(num):
  is_min = num == num.min()
  return['background-color: yellow' if i else " " for i in is_min]

data.style.apply(highlight_max, subset=pd.IndexSlice[:,['B', 'C', "D", "E"]])

data.style.apply(highlight_min, subset=pd.IndexSlice[:,['B', 'C', "D", "E"]])
```
